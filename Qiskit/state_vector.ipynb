```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.tools.visualization import plot_state_city
from qiskit import execute
from qiskit import BasicAer
```

```python
# qubit가 2개인 양자 레지스터 생성한다.
qbit=QuantumRegister(2, 'qbit')
```

```python
# 레지스터가 양자회로에서 동작할 수 있도록 해준다.
qcircuit=QuantumCircuit(qbit)
```

```python
# qubit 0에 아다마르곱(Gate H)을 사용해 중첩을 만어낸다.
qcircuit.h(0)
```

```python
# qubit 0을 control, qubit 1을 target으로 하는 CX(controlled-NOT) 게이트이다.
qcircuit.cx(0, 1)
```

```python
# 설계된 circuit를 가시화한다.
qcircuit.draw()
```

```python
# bakend 객체에 실행할 backend를 할당, 괄호 안에 들어가는 것은 backend의 환경이다.
bakend=BasicAer.get_backend('statevector_simulator')
```

```python
# exe 객체에 실행에 필요한 메소드를 할당한다.
exe=execute(qcircuit, bakend)
```

```python
# result 객체에 실행 결과를 할당한다.
result=exe.result()
```

```python
# 실행 결과를 상태 벡터로 가시화해준다.
output=result.get_statevector(qcircuit, decimals=2)
print(output)
```

```python
# 상태 벡터를 그래프로 가시화한다.
plot_state_city(output)
```
